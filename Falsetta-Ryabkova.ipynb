Our goal is to analyze the daily evolution of COVID-19 nationwide.
For this reason we create a variable "dates" and a function called daily_updates which counts the total number of cases for a certain variable of the data frame for each day.

```python
#we create a variable dates
dates=region["data"]

#obviously also whit this variable we're able to obtain the total number of days
print(len(dates)/num_reg)
```

```python
def daily_update(dates, values_attribute, col_name_attribute):
    
    """"
    
    We create a function in order to sum the values of the 21 regions in one national value for each day of survey.
    We need to consider that for each day we have there are 21 different values to account. 
    For this reason we'll use a counter which will be incremented of 21 in each cycle.
    The goal is to be able to plot the date of survey on the x-axes and total daily value for the specific column on y-axes.
    
    """
    
    days=[]
    tot=[]
    j=0
    
    for i in dates:
        if i[:10] not in days:
            days.append(i[:10])
            tot.append(sum(values_attribute[j:j+21]))
            j+=21
            
    result=pd.DataFrame({str(col_name_attribute):tot, "Date":days})        
    result.set_index("Date",inplace=True)
    
    return result
```

There is no need to represent cumulative national data for swabs (named swabs) graphically, since it doesn't seem to be statistically interesting.
We will use the cumulative distribution in order to obtain the daily number of swabs with the function daily_value.

```python
def daily_value(dates,value_attribute,col_name):      # we need to extrapolate daily swabs, deaths etc
    """
    We create 2 empty lists, one will contain the sum of the values for that specific variable, and the second will store the 
    calendar day.
    
    """
    
    daily_tot=[] 
    days=[]
      
    for i in dates:             # with this for cycle we append the calendar days to the list
        if i[:10] not in days:
            days.append(i[:10]) # we do not need to se the time but only the date
    
    for i in range(0,len(value_attribute)-1): 
        if i==0:
            daily_tot.append(value_attribute[i])
        daily_tot.append(value_attribute[i+1]-value_attribute[i])
        
        
    
    result=pd.DataFrame({str(col_name):daily_tot, "Date":days})        
    result.set_index("Date",inplace=True)
    
    return result
        
    
```

```python

daily_swabs=daily_value(dates,swabs['Swabs'],'Swabs')    # daily swabs obtained from the cumulative    

pd.set_option('display.max_rows', None)
daily_swabs  # we want to see all of the daily numbers in order to check possible anomalies
```
